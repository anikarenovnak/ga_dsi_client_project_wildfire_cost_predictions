```python
def lowering(df):
        lowered_columns = []
        columns = df.columns
        for item in columns:
            item = item.lower()
            item = item.replace(" ", "_")
            item = item.strip('_')
            lowered_columns.append(item)
        lowered_columns[0] = 'county' #fixes the county index issue.
        df.columns = lowered_columns
        return df
    
```

```python
umdf = lowering(umdf)
```

```python
county_list = umdf.area_name.unique()
year_list = umdf.year.unique()
```

```python
new_df = pd.DataFrame(columns = ['county_year','labor_force', 'employment', 'unemployment', 'unemployment_rate'])
```

```python
new_df
```

```python
for county in county_list:
    for year in year_list:
        try:
            mask_county = umdf.area_name == county
            mask_year  = umdf.year == year
            test_df = umdf[mask_county & mask_year]
            stringy = county+"_"+str(year)
            print (stringy)
            inserter = [stringy]
            inserter.append(round(np.mean(test_df.labor_force)))
            inserter.append(round(np.mean(test_df.employment)))
            inserter.append(round(np.mean(test_df.unemployment)))
            inserter.append(inserter[3]/inserter[1])
            print(inserter)
            new_df.loc[stringy] = inserter
            
        except:
            continue
```
