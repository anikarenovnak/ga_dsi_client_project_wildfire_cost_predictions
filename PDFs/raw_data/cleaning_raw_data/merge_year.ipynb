```python
# this notebook merges all the years of aligned raw data from calfire
```

```python
import pandas as pd
```

```python
df1 = pd.read_csv('./2016_fire_cause_by_county.csv')
```

```python
df2 = pd.read_csv('./2016_fire_size_by_county.csv')
```

```python
df3 = pd.merge(df1, df2, how = 'outer', on = 'county')
```

```python
df3.head()
```

```python
df4 = pd.read_csv('./2016_Redbook_AcresByCause_ByCounty_Edit.csv')
```

```python
df5 = pd.merge(df3, df4, how = 'outer', on = 'county')
df5.head()
```

```python
df6 = pd.read_csv('./2016_Redbook_AcresByVegType_ByCounty.csv')
```

```python
df7 = pd.merge(df5, df6, how = 'outer', on = 'county')
df7.head()
```

```python
df8 = pd.read_csv('./2016_Redbook_DollarsByCause_ByCounty.csv')
```

```python
df9 = pd.merge(df7, df8, how = 'outer', on = 'county')
df9.head()
```

```python
df10 = pd.read_csv('./2016_Redbook_DollarDamage_Size.csv')
```

```python
df11 = pd.merge(df9, df10, how = 'outer', on = 'county')
df11.head()
```

```python
df11.columns = [col.lower().replace(' ', '_') for col in df11.columns]
```

```python
df11.head()
```

```python
df11.to_csv(f'./2016_wildfire_statistics.csv', index = False)
```
