```python
# this notebook renames the column name to lower and replaces spaces with underscore
```

```python
import pandas as pd
```

```python
df_2010 = pd.read_csv('./2010_wildfire_statistics.csv')
```

```python
df_2010.head()
```

```python
df_2010.columns = [col.lower().replace(' ', '_') for col in df_2010.columns]
```

```python
df_2010.head()
```

```python
df_2010.to_csv(f'./2010_wildfire_statistics.csv', index = False)
```
